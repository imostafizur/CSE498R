```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
```

```python
from tensorflow.keras.models import load_model
model = load_model('Saved Model//brain_tumor.h5')
```

```python
# load image dataset
image = ('brain_tumor_dataset/yes/Y1.jpg')
```

```python
# Model summary
model.summary()
```

```python
image = ('brain_tumor_dataset//yes/Y243.JPG')
img = plt.imread(image)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img = tf.keras.preprocessing.image.load_img(image, target_size=(224, 224))
img = tf.keras.preprocessing.image.img_to_array(img)
img = np.expand_dims(img, axis=0)
img = img/255
prediction = model.predict(img)
print('Prediction:', prediction)
```

```python
# Prediction if else statement
if prediction > 0.5:
    print('The person has a brain tumor')
else:
    print('The person does not have a brain tumor')
```
